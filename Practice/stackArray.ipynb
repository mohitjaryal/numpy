# Stack two arrays vertically and horizontally using vstack() and hstack().

```python
import numpy as np
```

```python
arr1 = np.array([[1, 2, 3],
              [4, 5, 6]])
arr2 = np.array([[7, 8, 9],
              [10, 11, 12]])
print('Original array :',arr1)
print('Vstack :',np.vstack(arr1))
print('Hstack :',np.hstack(arr2))
```
