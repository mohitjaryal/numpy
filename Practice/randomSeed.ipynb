# Set a random seed (np.random.seed()) and generate the same random numbers again to check reproducibility.

```python
import numpy as np
```

```python
# Random seed
np.random.seed(4)

arr1 = np.random.randint(1,101, size=5)
print('First run :',arr1)

# Random seed
np.random.seed(4)

arr2 = np.random.randint(1,101,size=5)
print('Second run :',arr2)
```
