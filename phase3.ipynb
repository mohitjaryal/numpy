```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data structure: [restaurant_id, 2021, 2022, 2023]
sales_data = np.array([
    [1,150000, 120000, 250000, 100000], # PizzaHut
    [2,110000, 130000, 200000, 200000], # BurgerTime
    [3,100000, 200000, 290000, 220000], # ChillOut
    [4,300000, 190000, 300000, 210000], # PizzaCafe
    [5,700000, 100000, 700000, 230000], # Chai Point
])

print('==== Zomato Sales Analysis ====')
print('Sales data shape',sales_data.shape)
print('Sales data for 1st 3 reataurant :',sales_data[:3])
```

# Total sales per year

```python
print('Total sales per year :',np.sum(sales_data,axis=0))
yearly_total = np.sum(sales_data[:,1:],axis=0)
print('Yearly sum :',yearly_total)

# Minimum sales per restaurant
min_sales = np.min(sales_data[:,1:],axis=1)
print('Min sales per restaurant :',min_sales)
```
