# Learning numpy

```python
import numpy as np
```

## Creating arrays from list

```python
arr_1d = np.array([1,2,3,4,5])
print('1d array :',arr_1d)

arr_2d = np.array([[1,2,3],[4,5,6]])
print('2d array :',arr_2d)
```

## Lists vs array

```python
import time 
 
start = time.time()
py_pist = [i * 2 for i in range(10000000)] 
print('\nList operation time :',time.time() - start)


np_array = np.arange(10000000)
print('\nArrayoperation time :',time.time() - start)

```

## Crating arrays from scratch

```python
zeros = np.zeros((3,4))
print(zeros)

ones = np.ones((3,4))
print(ones)

full = np.full((2,3),4)
print(full)

random = np.random.random((2,3))
print(random)

sequence = np.arange(0,10,2)
print(sequence)
```

### Vector, Matrices and Tensor

```python
vector = np.array([1,2,3])
print('vector :\n',vector)

matrix = np.array([[1,2,3],
                   [4,5,6]])
print('matrix :\n',matrix)

tensor = np.array([ [[1, 2], [3, 4]],
                    [[5, 6], [7, 8]]])
print('tensor:\n', tensor)
```

### Properties of Array

```python
arr = np.array([[1,2,3],
                [4,5,6]])
print('Shape :', arr.shape)
print('Dimension :', arr.ndim)
print('Size :', arr.size)
print('Data type :', arr.dtype)
```

### Array Reshaping

```python
arr = np.arange(12)
print('Original array :\n', arr)

reshaped = arr.reshape((3,4))
print('Reshaped array :\n', reshaped)

flattend = reshaped.flatten()
print('Flattened array :\n', flattend)

ravel = reshaped.ravel()
print('Raveled array :\n', ravel)
```
